```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Drop duplicates if any
print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
# 🔁 Encoding categorical features (if any present in df)

# Check for categorical columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()
if categorical_cols:
    print("Categorical columns found:", categorical_cols)
    from sklearn.preprocessing import OneHotEncoder
    df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
    print("Encoding complete. Encoded DataFrame shape:", df_encoded.shape)
else:
    print("No categorical columns found to encode.")
```

```python
# Outlier analysis using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print("Shape after outlier removal:", df.shape)
```

```python
# Skewness and Kurtosis

from scipy.stats import kurtosis, skew
for col in df.columns:
    print(f"{col}: Skewness={skew(df[col]):.2f}, Kurtosis={kurtosis(df[col]):.2f}")
```

```python
# Apply log transformation to skewed columns
skewed_cols = [col for col in df.columns if abs(skew(df[col])) > 1]
for col in skewed_cols:
    df[col] = np.log1p(df[col])
```
